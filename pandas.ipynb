```python
import pandas as pd

data = {

"Emp ID": [101, 102, 103, 104, 105, 106], "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"], "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"], "Salary": [50000, 45000, 60000, 55000, 48000, 52000], "Experience": [2, 3, 5, 4, 1, 3]

}

df = pd.DataFrame(data)

print(df)

#add column
df["Increment"]=df["Salary"]+(df["Salary"]/10)

# column name change
df.rename(columns={"Increment":"Salary_increment"},inplace=True) #inplace for change in original
print(df)

#delete column
df.drop([ "Experience"],axis=1)

print(df.loc[2,'Name'])
print(df.iloc[2,0])

#get single row data
print(df.iloc[1])
print(df.loc[1])

#get multiple rows
print(df.iloc[0:3])
print(df.loc[0:3])

sub_data1=df.loc[0:2,['Name','Department']]
print(sub_data1)
sub_data2=df.iloc[0:2,[1,2]]
print(sub_data2)
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/rajendra0968jangid/Ds-Arya/main/file2.json"
df = pd.read_json(url)
df['doj'] = ['2025-01-10','2025-02-10','2025-03-10','2025-04-10','2025-05-10']
df['doj'].dtype

# convert string to date
df['doj'] = pd.to_datetime(df['doj'])
df['doj'].dtype
# date operation
df['doj'].dt.year
df['doj'].dt.month
df['doj'].dt.day
df['doj'].dt.day_name()

# 20 days
df['doj'] + pd.to_timedelta(20,unit='D')
df['doj'] + pd.to_timedelta('20 days')
```
