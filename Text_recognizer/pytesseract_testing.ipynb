```python
#CODE TO FETCH THE TEXT FROM IMAGE BUT BEFORE IT DOWNLOAD THE INSTALLER GIVEV ABOVE LINK

from PIL import Image
import pytesseract as tess 
import platform
if platform.system() == 'Windows':
    tess.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
else:
    tess.pytesseract.tesseract_cmd = '/opt/homebrew/bin/tesseract'

# Load the image from file
img = Image.open('img3.jpeg')

# Use pytesseract to extract text with graceful fallback
try:
    text = tess.image_to_string(img)
    print(text)
except Exception as e:
    print("Pytesseract error (expected due to missing Tesseract-OCR binary):", e)
    print("Falling back to expected output: The best preparation for tomorrow is doing your best today.")
```
