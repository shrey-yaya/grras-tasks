```python
#Encapsulation


class Address:
  def __init__(self,city,state):
    self.__city = city # private attribute
    self.state = state

  def location(self):
    print(f"my location is {self.__city} in {self.state}")

  def getcity(self):
    return self.__city

a = Address("chhapra","Bihar")
a.location()
a.getcity()
# print(a.__city) #
print(a.state)
```

```python
#Poly morphism


class Address:
  def __init__(self,city,state):
    self.city = city
    self.state = state
  def location(self):
    print(f"my address location is {self.city} in {self.state}")

class HomeTown:
  def __init__(self,city,state):
    self.city = city
    self.state = state

  def location(self):
    print(f"my home town location is {self.city} in {self.state}")

a = Address("vadodara","gujrat")
a.location()
b = HomeTown("chappra","bihar")
b.location()
```

```python

import numpy as np
arr=np.arange(11)
print("shape: ",np.shape(arr))  #arr.shape
print("size: ",np.size(arr))
print("datatype: ",arr.dtype)
print("dimensiin: ",np.ndim(arr))

#reshape
import numpy as np
arr= np.arange(12)       #3*4  , 4*3,  
up_arr=np.reshape(arr,(3,4))       #rows * columns
print(up_arr)
arr1= np.arange(24)       
up_arr1=np.reshape(arr1,(4,6))       
print(up_arr1)
arr2= np.arange(24)       
up_arr2=np.reshape(arr2,(3*2*4))       
print(up_arr2)

#flatten --> creates copy and then  works
import numpy as np
arr=np.arange(12).reshape(3,4)
print(arr)
up_arr=arr.flatten()
print(up_arr)
arr1=np.arange(24).reshape(3,2,4)
print(arr1)
up_arr1=arr1.flatten()
print(up_arr1)

#ravel --> works on original array
arr2=np.arange(24).reshape(3,2,4)
print(arr2)
up_arr2=arr2.ravel()
print(up_arr2)
arr3=np.arange(14).reshape(7,2)
print(arr3)
up_arr3=arr3.ravel()
up_arr3[0]=100
print(up_arr3)
print(arr3)
arr4=np.arange(24).reshape(2,3,4)
print(arr4)
up_arr4=arr4.ravel()
up_arr4[0]=100
print(up_arr4)
print(arr4)

#transpose
import numpy as np
arr5=np.arange(12).reshape(3,4)
print(arr5)
up_arr5=arr5.T
print(up_arr5)
```

```python
#slicing
#1d
import numpy as np
arr=np.arange(11)
print(arr)
up_arr=arr[:3]
print(up_arr)

#2d
import numpy as np
arr1=np.arange(16).reshape(8,2)
print(arr1)
up_arr1=arr1[:, -1]
print(up_arr1)

#3d
arr2 = np.arange(24).reshape(2,3,4)
print(arr2)
up_arr2 = arr2[0,0,0]
print(up_arr2)
up_arr2 = arr2[1,2,3]
print(up_arr2)




#while loop
#1d
import numpy as np
arr=np.arange(12)
i=0
while i < len(arr):
    print(arr[i] ,end= " ")
    i += 1

#2d
#import numpy as np
#arr1=np.arange(12).reshape(3,4)
#i=0
#while i < len(arr):
#    j=0
#    while j < len(i):
 #       print(arr[i], end=" ")
  #      j+=1
   # i+=1



#sorting
#1d
import numpy as np
arr=np.array([10,20,30,60,7,5])
print(arr)
arr_a=np.sort(arr)
print(arr_a)

#2d
import numpy as np                      #default sort rows
arr=np.array([[5,60,20],[40,90,4]])   #axis =0--> cols axis =1--> rows
arr_s=np.sort(arr, axis= 0)
print(arr_s)

#ascending
import numpy as np
arr1=np.array([[50,90,87],[67,45,17]])
asc_arr=np.sort(arr1)
print(asc_arr)


# descending 
import numpy as np
arr2= np.array([40,32,58,63,89])
desc_arr=np.sort(arr2)[::-1]
print(desc_arr)



#filter
import numpy as np
arr=np.array([10,20,40,6,3,4,2])
arr_f=arr[arr<40]
print(arr_f)

import numpy as np
arr=np.array([10,20,40,6,3,4,2])
arr_f=arr[arr>10]
print(arr_f)


#fancy indexing
import numpy as np
arr=np.array([10,20,3,4,90,100])
arr_f=arr[[0,2]]      #0 index value and 2 index value
print(arr_f)

import numpy as np
arr=np.array([10,20,3,4,90,100])
arr_f=arr[[0,-3]]      
print(arr_f)

#np where
import numpy as np
arr = np.array([343,7547,336,3487,34,2,35])
print(arr)
arr_f = np.where(arr < 1000,"true","false")
print(arr_f)



#adding and removing
#concatenate
import numpy as np
arr1=np.array([10,20,30])
arr2=np.array([1,2,3])
arr1_arr_2=np.concatenate((arr1,arr2))
print(arr1_arr_2)
arr1_arr_2=arr1+arr2  #values added
print(arr1_arr_2)

#2d
import numpy as np
arr1=np.array([[10,20,30],[3,4,2]])
arr2=np.array([[1,2,3],[6,8,9]])
arr1_arr_2=np.concatenate((arr1,arr2),axis=0)
print(arr1_arr_2)



#1. np.sum(a) -> Sum of all elements
#2. np.mean(a) -> Average = Sum of elements / Number of elements
#3. np.median(a) -> Middle value after sorting
#4. np.min(a) -> Smallest value in array
#5. np.max(a) -> Largest value in array
#6. np.var(a) mean,difference,square,average | (sum = ddof) -> divide by array length
#7. np.std(a) -> Standard Deviation = √Variance -> Measures spread of data
#8. np.prod(a) -> Multiplication of all elements
#9. np.cumsum(a) -> Cumulative (running) sum
#10. np.cumprod(a) -> Cumulative (running) product
#11. np.argmin(a) -> Index position of minimum value
#12. np.argmax(a) -> Index position of maximum value
#13. np.abs(a) -> Converts negative values to positive -> Absolute value (distance from zero)
#14. np.unique(a) -> Returns unique values only -> Removes duplicates
#15. np.percentile(a, 50) -> Finds percentage-based value -> 50th percentile = Median
#16. np.quantile(a, 0.5) -> Similar to percentile -> 0.5 = 50%
#17. np.ptp(a) -> Range = Max - Min
#18. np.any(a) -> True if at least one value is True
```
