PRIMITIVE DATATYPES--> integers, float, strings, boolean, none

```python
#integers
age=25
print(type(age))
type(age)

#floats
percent=87.5
print(type(percent))

print(10+20)
3.5+4.7
4+3.5

#strings
name="mark"
print(name,type(name))

full_name="mark wood"
print(full_name,type(full_name))

#boolean-->True and False
vall=True
print(vall,type(vall))

val=False
print(val,type(val))

#None
val3=None
vall=True
print(val3,type(val3))
```
